# Do the conversion

```python
mentions = []
relations = []

# READ OFF ENTITIES, ENTITY MENTIONS, and COREF
# create a full mention annotation and coref relations for each entity
for e in annotation_soup.find_all('entity'):
    print 50*'-'
    entity = e.attrs
    print entity
    e_mentions = []
    for em in e.find_all('entity_mention'):
        start, end = int(em.extent.charseq['start'])-1, int(em.extent.charseq['end'])
        if (start >= end): 
            print em
        seq = em.extent.charseq.text
#         print em
        mention = {'ann-type':'node',
                   'ann-span':charspan2tokenspan(int(em.extent.charseq['start']), 
                                                 int(em.extent.charseq['end'])),
                   'ann-uid':em['id'],
                   'type':e['type'],
                   'subtype':e['subtype'],
                   'entity-uid':e['id'],
                   'doc-uid': doc_id,
                   'class':e['class']}
#         print mention
        e_mentions.append(mention)
    e_mentions.sort(key=lambda x:x['ann-span'][1])
    for i, em1 in enumerate(e_mentions):
        for j, em2 in enumerate(e_mentions[i+1:]):
            d = em2['ann-span'][1] - em1['ann-span'][0]
            if d <=0:
                print em1['ann-span'], tokens[em1['ann-span'][0]:em1['ann-span'][1]]
                print em2['ann-span'], tokens[em2['ann-span'][0]:em2['ann-span'][1]]
                print '-'*80
            coref = {'ann-type':'edge',
                     'ann-left':em1['ann-span'][1],  #             |-------------|
                     'ann-right':em2['ann-span'][0], # [e1 ... [mention]] ... [[e2] ... mention]
                     'ann-dist':d,
                     'ann-left-uid':em1['ann-uid'],
                     'ann-right-uid':em2['ann-uid'],
                     'type': 'sameAs',
                     'doc-uid':doc_id}
            relations.append(coref)
    mentions.extend(e_mentions)
```

```python
import json
with open('../mentions.json', 'w') as f:
    f.write(unicode(json.dumps(mentions)))
with open('../tokens.json', 'w') as f:
    f.write(unicode(json.dumps([ str(t) for t in tokens ])))
```

```python
def relation_nominals(relation):
    e1 = em_dict[r['ann-left-uid']]
    e2 = em_dict[r['ann-right-uid']]
    return "(d={}) :: ({}:{}) - {} - ({}:{})".format(r['ann-dist'],
                                     e1['type'],
                                     " ".join(tokens[e1['ann-span'][0]:e1['ann-span'][1]]),
                                     relation['type'],
                                     e2['type'],
                                     " ".join(tokens[e2['ann-span'][0]:e2['ann-span'][1]]))
```
